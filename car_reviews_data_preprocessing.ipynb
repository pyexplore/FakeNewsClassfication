```python
import pandas as pd
import numpy as np

import re
```

```python
input_file_path = 'data/Edmunds/Scraped_Car_Review_ford.csv'
output_file_path = 'data/cleaned/Scraped_Car_Review_ford.csv'

# Read the file
with open(input_file_path, 'r', encoding='utf-8') as reader:
    count = 0
    lines = ""
    for line in reader:
        
        match_obj = re.search('\d+, on \d+/\d+/\d+ \d+:\d+ [P|A]M \(...\),', line)
        if match_obj:
            lines += '\n' + line.rstrip('\n')
        else:
            lines += " " + line.rstrip('\n')

# Write the cleaned lines to a new file
with open(output_file_path, 'w', encoding='utf-8') as writer:
    writer.writelines(lines)
```
